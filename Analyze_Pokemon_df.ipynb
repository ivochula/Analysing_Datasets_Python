```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\Projetos\Datasets de treino\Pokemon\Poke.csv")
df.head()
```

```python
df_group = df.groupby(["Type 1"]).count().sort_values("Total")
df_group
```

```python
type1 = df_group.index
count_type1 = df_group.Total
```

```python
labels = type1
values = count_type1

# Resize Plot
plt.figure(figsize=(5,3), dpi=300)

plt.bar(labels, values, color = 'g', label = "Count",  edgecolor = "black", width = .6)
plt.title( "Number of Primary Types of Pokemon", fontdict = {'fontsize': 14} )
plt.xticks(rotation = 90)
plt.tight_layout()

plt.legend()

plt.show()
```

### We can see by this graph above that the number one type of pokemons is Water followed closely by normal type. ###
### Flying pokemons are the ones with least species. ###

```python
df_legendary = df.groupby(["Legendary"]).count().sort_values("Type 1")
```

```python
legendary_count = df_legendary.Total
```

```python
labels = ["Legendary","Normal"]
sizes = legendary_count
colors = ['red', 'green']

plt.pie(sizes, labels = labels, colors = colors, autopct='%1.2f%%', radius = 1, 
        rotatelabels = 0, textprops=dict(color="w", fontsize=12, weight="bold"))
plt.title( "Amount of Legendary Pokemons", fontdict = {'fontsize': 18}, bbox={'facecolor':'0.9', 'pad':6} )

plt.legend( loc = "lower right") 

plt.show()
```

```python
df_power = round(df.groupby(["Type 1"]).mean().sort_values("Attack", ascending = False))
```

```python
df_power
```

```python
labels = df_power.index
values = df_power.Attack

plt.bar(labels, values, color = 'g', label = "Attack Power",  edgecolor = "black", width = .6)
plt.title( "Strongest Attacking Pokemon Type", fontdict = {'fontsize': 14} )
plt.xticks(rotation = 90)
plt.tight_layout()

plt.legend()

plt.show()
```

```python
df_defense = round(df.groupby(["Type 1"]).mean().sort_values("Defense", ascending = False))
```

```python
df_defense
```

```python
labels = df_defense.index
values = df_defense.Defense

plt.bar(labels, values, color = 'b', label = "Defense Power",  edgecolor = "black", width = .6)
plt.title( "Strongest Defending Pokemon Type", fontdict = {'fontsize': 14} )
plt.xticks(rotation = 90)
plt.tight_layout()

plt.legend()

plt.show()
```
